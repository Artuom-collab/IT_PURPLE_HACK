```python
import pandas as pd
import numpy as np
import tqdm

# Функция для оптимизации использования памяти
def reduce_mem_usage(df, cat_cols=[]):
    start_mem = df.memory_usage().sum() / 1024**2

    for col in tqdm.tqdm(df.columns):  # Исправлено для отображения прогресса
        col_type = df[col].dtype.name

        if col_type not in ['object', 'category', 'datetime64[ns, UTC]']:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    df = df.astype({cat_col: "category" for cat_col in cat_cols})

    end_mem = df.memory_usage().sum() / 1024**2
    return df

# Модель
df = pd.read_pickle("C:\\Users\\92672\\Downloads\\df.pkl")

# Применение функции оптимизации памяти
df = reduce_mem_usage(df)
df

#to_drop_cols = ['feature642', 'feature756']
#df = df.drop(columns=to_drop_cols, axis=1)
```

```python
import pandas as pd

# Теперь исполним ваш исходный код с функцией поиска дублирующихся столбцов
def find_duplicate_feature_columns(df, feature_prefix='feature'):
    duplicate_columns_pairs = []
    feature_columns = [col for col in df.columns if col.startswith(feature_prefix)]
    for i in range(len(feature_columns)):
        for j in range(i + 1, len(feature_columns)):
            col1, col2 = feature_columns[i], feature_columns[j]
            if df[col1].equals(df[col2]):
                duplicate_columns_pairs.append((col1, col2))
    return duplicate_columns_pairs

# Найти дублирующиеся признаки фич
duplicates = find_duplicate_feature_columns(df, feature_prefix='feature')

# Вывести результат
print("Пары дублирующихся признаков:")
for pair in duplicates:
    print(f"{pair[0]} и {pair[1]}")

# Создание DataFrame с дублирующимися столбцами
if duplicates:
    df_duplicates = pd.DataFrame(duplicates, columns=['Feature_1', 'Feature_2'])
    print(df_duplicates)
else:
    print("Дублирующихся признаков не найдено.")

# Сохраняем результат в файл в формате .pkl
if duplicates:
    df_duplicates.to_pickle("duplicates.pkl")
    # Вывести индексы совпадающих столбцов
    print("Числа фич, признаки которых совпадают:")
    print(df_duplicates)
else:
    print("Нет данных для сохранения.")
```
